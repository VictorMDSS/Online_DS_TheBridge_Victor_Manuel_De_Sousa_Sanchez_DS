```python
df_california = pd.read_csv("./data/california_cities.csv")
print(df_california.info())
print("\n")
print(df_california.head())
df_california

```

```python
df_california.dropna(inplace=True)
df_california.info()
```

```python
df_ordenado = df_california.sort_values(by= "population_total", ascending= False)
df_5_mas_pobladas = df_ordenado.head(5)
df_5_mas_pobladas
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
fig, ax = plt.subplots()

categorias = df_5_mas_pobladas["city"].to_list()
frecuencia = df_5_mas_pobladas["population_total"].to_list()
colores = ["blue", "gray", "brown", "pink", "orange"]

ax.bar(categorias, frecuencia, color= colores)
ax.set_title("Ciudades ms pobladas de California")
ax.set_xlabel("ciudades")
ax.set_ylabel("Habitantes (miles)")
ax.set_ylim(0, 5000000);

# Anotaciones
for i in range(len(df_5_mas_pobladas)):
    ax.annotate(f'{df_5_mas_pobladas["population_total"].values[i]:,}', 
                (df_5_mas_pobladas["city"].values[i], df_5_mas_pobladas["population_total"].values[i]),
                textcoords="offset points", 
                xytext=(0,5), 
                ha='center')
```

```python
font_dict = {
    "size":18,
    "family": "Bradley Hand itc",
    "color": "steelblue"
}

plt.hist(df_california["area_total_km2"], bins= 50, color= "steelblue", label="Hist Area")
plt.xlabel("Superficie (Km2)", fontdict= font_dict)
plt.ylabel("Numero de ciudades", fontdict=font_dict)
plt.legend();
```

```python
plt.scatter(df_california["area_water_km2"], df_california["population_total"], c= "g", s = 20, alpha = 0.5)
max_population = df_california["population_total"].max()
max_water = df_california["area_water_km2"].max()

plt.annotate(xy= (85,max_population), xytext=(45,2550000), text="Outliers", arrowprops= dict(facecolor= "black",shrink = 0.02))
plt.annotate(xy= (120,1400000), xytext=(62,2550000), text="", arrowprops= dict(facecolor= "black",shrink = 0.003))
plt.title("Poblacin por Kilometro cuadrado de agua")
```

```python
fig, ax = plt.subplots()
ax.boxplot(df_california["elevation_m"], vert=False, widths= 0.2, patch_artist=True)
ax.set_title("Diagrama de Caja de la Altitud de las Ciudades de California")
ax.set_xlabel("Altitud (m)")
```

```python
fig, ax = plt.subplots(2,2, figsize = (15,11))

fig.subplots_adjust(hspace=0.5)


# Grafico de barras
categorias = df_5_mas_pobladas["city"].to_list()
frecuencia = df_5_mas_pobladas["population_total"].to_list()
colores = ["blue", "gray", "brown", "pink", "orange"]

ax[0,0].bar(categorias,frecuencia, color = colores)
ax[0,0].set_title("Ciudades ms pobladas de California")
ax[0,0].set_xlabel("Ciudades")
ax[0,0].set_ylabel("Habitantes (miles)")
ax[0,0].set_ylim(0, 5000000);

# Anotaciones Grafico de barras
for i in range(len(df_5_mas_pobladas)):
    ax[0,0].annotate(f'{df_5_mas_pobladas["population_total"].values[i]:,}', 
                (df_5_mas_pobladas["city"].values[i], df_5_mas_pobladas["population_total"].values[i]),
                textcoords="offset points", 
                xytext=(0,5), 
                ha='center')
    
#Histograma de Superficie
font_dict = {
    "size":18,
    "family": "Bradley Hand itc",
    "color": "steelblue"
}

ax[0,1].hist(df_california["area_total_km2"], bins= 50, color= "steelblue", label="Hist Area")
ax[0,1].set_xlabel("Superficie (Km2)", fontdict= font_dict)
ax[0,1].set_ylabel("Numero de ciudades", fontdict=font_dict)
ax[0,1].legend();


# Diagrama de Dispersion del Agua

ax[1,0].scatter(df_california["area_water_km2"], df_california["population_total"], c= "g", s = 20, alpha = 0.5)
max_population = df_california["population_total"].max()
max_water = df_california["area_water_km2"].max()

ax[1,0].annotate(xy= (85,max_population), xytext=(45,2550000), text="Outliers", arrowprops= dict(facecolor= "black",shrink = 0.02))
ax[1,0].annotate(xy= (120,1400000), xytext=(62,2550000), text="", arrowprops= dict(facecolor= "black",shrink = 0.003))
ax[1,0].set_title("Poblacin por Kilometro cuadrado de agua")

# Diagrama de Caja de la Altitud de las Ciudades de California

ax[1,1].boxplot(df_california["elevation_m"], vert=False, widths= 0.2, patch_artist=True)
ax[1,1].set_title("Diagrama de Caja de la Altitud de las Ciudades de California")
ax[1,1].set_xlabel("Altitud (m)")



```
